```python
# function to get more than 100 songs from playlist
def get_playlist_tracks(user_id,playlist_id):
    results = sp.user_playlist_tracks(user_id,playlist_id)
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    return tracks
```

```python
#apply the function 
summerhits2=get_playlist_tracks("spotify", "3LCQR56VQ40OrIdXbubzp3")
```

```python
#function to retrieve ids from the json 
def get_info_from_tracks(track_list):
    track_ids=[]
    for track in track_list:
        track_ids.append(track['track']['id'])
    return track_ids
```

```python
#apply the function 
pl_track_ids=get_info_from_tracks(summerhits2)
```

```python
#check the results 
len(pl_track_ids)
```

```python
# as we received a max retries error, lets try splitting our request 
#into blocks of 50 records. if this doesnt work, theres a bad song id
# found this solution on stackoverflow
# https://stackoverflow.com/questions/312443/how-do-you-split-a-list-into-evenly-sized-chunks

def sections(lst, n):
    """Yield successive n-sized sections from list of ids(lst)."""
    for i in range(0, len(lst), n):
        yield lst[i:i + n]
```

```python
#use a for loop to run through the ids, collect the audio features
#in sections of 50 per call 
pl_audiofeats = []
for section in list(sections(pl_track_ids, 50)):
        pl_audiofeats.append(sp.audio_features(section))
```
