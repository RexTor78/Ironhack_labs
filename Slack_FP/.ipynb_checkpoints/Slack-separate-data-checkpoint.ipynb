```python
date_time = []
user = []
text = []
file_content = []
reactions = []
reactions_user=[]
replies_user = []
reply_users=[]
channel=[]
replies_time=[]
link=[]
file_url=[]
#for l in lab_questions:
for l in lab_questions:    #lab_questions[0][0]['ts']
    for mes in l:
        #channel.append(l)
    
        date_time.append(mes["ts"])
        text.append(mes['text'])
        user.append(mes["user"])
        file_content.append("files" in mes.keys())
        if 'files' in mes.keys():
            file_url.append(mes['files'][0]['url_private_download'])
        else:
            file_url.append(0)
        if 'attachments' in mes.keys():
            if 'title_link' in 'attachments':
                link.append(mes['attachments'][0]['title_link'])
            else:
                link.append(0)
        else:
            link.append(0)
        if 'reply_users_count' in mes.keys():
            reply_users.append(mes['reply_users_count'])
        else: 
            reply_users.append(0)

        if 'replies' in mes.keys():
            replies_time.append(mes['replies'][0]['ts'])
            replies_user.append(mes['replies'][0]['user'])
        else: 
            replies_time.append(0)
            replies_user.append(0)

        #if 'reactions' in mes.keys():
            #count=0
            #for reaction in mes["reactions"]:
                #count+=reaction['count']
                #reactions.append(count)
        #else: 
            #reactions.append(0)
        if 'reactions' in mes.keys():
            reactions_user.append(mes['reactions'][0]['users'][0])
        else:
            reactions_user.append(0)
            
lab_questions = pd.DataFrame({"daytime":date_time,
                    "user": user,'text':text,'file_content':file_content,
                    'replies_time':replies_time,'reply_users':reply_users,'replies_user':replies_user,
                    'file_url':file_url,'link':link,
                    'reactions_user':reactions_user})
```

```python
def times ():
    for i in lchannel:
        
        i['daytime']=pd.to_numeric(i['daytime'])
        i['replies_time']=pd.to_numeric(i['replies_time'])
        i["daytime"] = i["daytime"].apply(lambda x: round(x))
        i["replies_time"] = i["replies_time"].apply(lambda x: round(x))
        def time_conversor(x):
            return datetime.utcfromtimestamp(x).strftime('%d-%m-%Y %H:%M')
        i["dayt"] = i["daytime"].apply(lambda x: time_conversor(x))
        i["rep_time"] = i["replies_time"].apply(lambda x: time_conversor(x))
        i["dayt"]=i["dayt"].apply(lambda x: datetime.strptime(x,'%d-%m-%Y %H:%M'))
        i['date'] = pd.to_datetime(i['dayt']).dt.date
        i['time'] = pd.to_datetime(i['dayt']).dt.time
        i['rep_time']=i['rep_time'].astype(str)
        #df_final["dayt"]=df_final["dayt"].apply(lambda x: datetime.strptime(x,'%H:%M'))
        i['rep_time'] = pd.to_datetime(i['rep_time']).dt.time
        i.drop(['dayt','daytime','replies_time'], axis=1, inplace=True)
```

```python
users=json.load(open('slack/users.json'))
user_id=[]
name=[]
for user in users:
    user_id.append(user['id'])
    if 'real_name' in user.keys():
        name.append(user['real_name'])
    else:
        name.append(0)

users=dict(zip(user_id,name))

```

```python
def user():
    for u in lchannel:
        
        id_list=list(u['user'])
        u['user']=[users[item] for item in id_list]
        rea_list=list(u['reactions_user']) 
        for i, item in enumerate(rea_list):
            if item != 0:
                rea_list[i]=users[item]
        u['reactions_user']=rea_list
        rep_list=list(u['replies_user'])
        for x, item in enumerate(rep_list):
            if item!=0:
                rep_list[x]=users[item]
        u['replies_user']=rep_list
```

```python
date_time = []
user = []
text = []
file_content = []
reactions = []
reactions_user=[]
replies_user = []
reply_users=[]
channel=[]
replies_time=[]
link=[]
file_url=[]
for c in channels:
    files = [f for f in listdir("slack/" +c) if isfile(join("slack/" +c, f))]

    for file in files:
        data=json.load(open( "slack/"+c +"/" + file,encoding='utf-8'))

        for message in data:
            channel.append(c)
            date_time.append(message["ts"])
            text.append(message['text'])
            user.append(message["user"])
            file_content.append("files" in message.keys())
            
            if 'files' in message.keys():
                file_url.append(message['files'][0]['url_private_download'])
            else:
                file_url.append(0)
            if 'attachments' in message.keys():
                if 'title_link' in 'attachments':
                    link.append(message['attachments'][0]['title_link'])
                else:
                    link.append(0)
            else:
                link.append(0)
            if 'reply_users_count' in message.keys():
                reply_users.append(message['reply_users_count'])
            else: 
                reply_users.append(0)
                    
            if 'replies' in message.keys():
                replies_time.append(message['replies'][0]['ts'])
                replies_user.append(message['replies'][0]['user'])
            else: 
                replies_time.append(0)
                replies_user.append(0)
            
            if 'reactions' in message.keys():
                count=0
                for reaction in message["reactions"]:
                    count=count+reaction['count']
                reactions.append(count)
            else: 
                reactions.append(0)
            if 'reactions' in message.keys():
                reactions_user.append(message['reactions'][0]['users'][0])
            else:
                reactions_user.append(0)
```
